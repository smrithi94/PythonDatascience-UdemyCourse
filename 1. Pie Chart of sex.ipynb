```python
import matplotlib.pyplot as plt
import sqlite3 
import pandas as pd
import numpy as np

df = pd.read_csv("adult.data.txt", header=None, delimiter=', ', engine='python', names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]);

total = {"Male": 0, "Female": 0}

for index, row in df.iterrows():
    total[row["sex"]] =  total[row["sex"]] + 1

def format(pct):
    return "{:.1f}%".format(pct)

plt.rcParams['figure.figsize'] = [10, 5]
plt.pie([total["Male"], total["Female"]], labels = None, shadow = False, autopct=lambda pct: format(pct), textprops=dict(color="w"))
plt.axis('equal')
plt.title('Sex of adults in the Data Set', pad=25)

legendEntries = []

legendEntries.append('Male - ' + str(total["Male"]) + ' records')
legendEntries.append('Female - ' + str(total["Female"]) + ' records')
plt.legend(legendEntries)

plt.show()
```
